```python
# Importing all libraries required in this 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Reading data from remote link
url = "http://bit.ly/w-data"
s_data = pd.read_csv(url)
print("Data imported successfully")
s_data.head(24)

# Create a DataFrame
df = pd.DataFrame(s_data, columns=["Hours", "Scores"])

# Create the linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(df[["Hours"]], df["Scores"])
# Print the coefficients
print(f"Coefficients: {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")

# Predict the score for a student who studies for 9.25 hours
predicted_score = model.predict([[9.25]])[0]
# Print the predicted score
print(f"Predicted score for 9.25 hours of study: {predicted_score:.2f}")

# Plot the data and the regression line
plt.scatter(df["Hours"], df["Scores"])
plt.plot(df["Hours"], model.predict(df[["Hours"]]))
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.title("Score vs. Hours Studied")
plt.show()
```
